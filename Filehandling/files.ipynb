```python

# File operations in Python involve reading from and writing to files
# Python provides built-in functions to handle file operations efficiently

def read_file(filename):
    """Read and return the contents of a file"""
    with open(filename, 'r') as file:
        return file.read()

def write_file(filename, content):
    """Write content to a file"""
    with open(filename, 'w') as file:
        file.write(content)

def append_to_file(filename, content):
    """Append content to an existing file"""
    with open(filename, 'a') as file:
        file.write(content)

def read_lines(filename):
    """Read file and return list of lines"""
    with open(filename, 'r') as file:
        return file.readlines()

def file_exists(filename):
    """Check if a file exists"""
    import os
    return os.path.exists(filename)
```

```python
def write_file(file_name, content):
    with open(file_name, 'w') as file:
        file.write(content)
```

```python
write_file('examplefile.txt', '''Hello every one one 
           
           This is the best place 
One of the best course ''')

write_file('ex1.py', read_file("examplefile.txt"))
print(read_file('ex1.py').strip())
```
