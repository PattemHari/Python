Oparetors
- Arithmetic Operators: +, -, %, /, //, *. **
- Assignment Operators: =, +=, -=, /=, **=, *=, //=
- Comparison Operators: ==, >, <, >=, <=, !=
- Logical Operators: AND, OR, NOT

```python
a = 20
b = 10
floor_division = a // b
division = a / b
print(type(floor_division))
print(type(division))
print("The floor division of a and b is", floor_division)
print("The division of a and b is", division)
```
