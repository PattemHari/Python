Condition Statements:
- if
- elif
- else

```python
print("Hello, World!")
```

```python
value1 = input("Enter the first value")
value2 = input("Enter the second value")

# using only if
if value1 > value2:
    print(value1, "is greater compared to ", value2)
```

```python
#Using if
if value1 < value2:
    print(value2, "is greater compared to ", value1)
```

```python
if value1 == value2:
    print("Both values are equal")
```

```python
# using if else
if value1 > value2:
    print(value1, "is greater compared to ", value2)
else:
    print(value2, "is greater compared to ", value1)
```

```python
# using if elif else
value1 = 20
value2 = 20
if value1 > value2:
    print(value1, "is greater compared to ", value2)
elif value1 < value2:
    print(value2, "is greater compared to ", value1)
else:
    print("Both values are equal")

#we can also use nested if else statements to compare the values
```
