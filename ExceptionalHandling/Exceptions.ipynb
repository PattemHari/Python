```python
y = 0
try:
    x = 1 / y
except ZeroDivisionError as e:
    print("divide by zero", e)
except (TypeError, ValueError):
    print("bad input")
else:
    print("success", x)
finally:
    print("always executes")
```

```python
# Definitions:
# try: block where you put code that might raise exceptions
# except: one or more handlers for specific exception types
# else: runs if try block did not raise any exception
# finally: runs always, whether exception occurred or not

y = 1 # existing variable

try:
    x = 1 / y
except ZeroDivisionError as e:
    print("divide by zero", e)  # handle division by zero
except (TypeError, ValueError) as e:
    print("bad input", e)  # handle invalid types/values
else:
    print("success", x)  # runs when no exception occurred
finally:
    print("always executes")  # always runs
```

```python
# you can raise an exception explicitly with `raise`
if y == 0:
    raise ZeroDivisionError("y is zero, throwing exception")
```

```python
try:
    x = int(input("Enter a number"))
    print("No error occured.......")
except ValueError as ex:
    print("error occured : ",ex)
except Exception as ex1:
    print("error : ",ex1)
else:
    print(f"the value you entered {x}")
```
