```python
#Dicn
# Dictionaries in Python are key-value data structures
# They are mutable, unordered (in Python 3.6+ they maintain insertion order), and allow fast lookups

# Creating dictionaries
empty_dict = {}
another_empty_dict = dict()

# Dictionary with initial values
student = {
    "name": "John",
    "age": 20,
    "grade": "A",
    "courses": ["Math", "Physics", "Chemistry"]
    }

    # Accessing dictionary values
print(student["name"])  # Output: John
print(student.get("age"))  # Output: 20

# Adding new key-value pairs
student["gpa"] = 3.8

# Modifying existing values
student["grade"] = "A+"

# Removing key-value pairs
del student["courses"]
# or use pop()
student.pop("gpa")

# Dictionary methods
print(student.keys())  # Output: dict_keys(['name', 'age', 'grade'])
print(student.values())  # Output: dict_values(['John', 20, 'A+'])
print(student.items())  # Output: dict_items([('name', 'John'), ('age', 20), ('grade', 'A+')])

# Iterating through dictionary
for key, value in student.items():
    print(f"{key}: {value}")
```

```python
students = {"name": "Hari", "age": 25, "gender": "Male"}
print(students["name"])
print(students.get("age"))
print(students.get("HDbhj"))  # This will return None
print(students["zjbvf"])  # This will raise a KeyError
```

```python
student_copy = students
print(student_copy) 
print(students)
students["name"] = "Srihari"
print(student_copy) # This will also show the updated name because student_copy is referencing the same dictionary as students
print(students)
```

```python
student_info = students.copy()  # This creates a shallow copy of the dictionary
print(student_info)  # Output: {'name': 'Srihari', 'age': 25, 'gender': 'Male'}

students["name"] = "Hari"
print(student_info)  # Output: {'name': 'Srihari', 'age': 25, 'gender': 'Male'}
print(students)  # Output: {'name': 'Hari', 'age': 25, 'gender': 'Male'}
# The copy() method creates a shallow copy of the dictionary, which means that it creates a new dictionary object but does not create copies of the objects that the original dictionary references. In this case, since the values in the dictionary are immutable (strings and integers), the shallow copy is sufficient. However, if the dictionary contained mutable objects (like lists or other dictionaries), changes to those objects in the original dictionary would also affect the copied dictionary.
```

```python
#Disctionry comprehension
squared_numbers = {x: x**2 for x in range(1, 6)}
print(squared_numbers)  # Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
# use if else elif in disctionry comprehension
number_type = {x: "Even" if x % 2 == 0 else "Odd" for x in range(1, 11)}
print(number_type)  # Output: {1: 'Odd', 2: 'Even', 3: 'Odd', 4: 'Even', 5: 'Odd', 6: 'Even', 7: 'Odd', 8: 'Even', 9: 'Odd', 10: 'Even'} 
```

```python
nums = [2, 2, 2, 3, 3, 4, 4, 5]
counter = {}
for i in nums:
    if i in counter:
        counter[i] += 1
    else:
        counter[i] = 1
print(counter)  # Output: {2: 3, 3: 2, 4: 2, 5: 1}
#use list comprehension to count the frequency of each number in the list
counter_comp = {i: nums.count(i) for i in set(nums)}
print(counter_comp)  # Output: {2: 3, 3: 2, 4: 2, 5: 1}
```

```python
nums2 = [1, 2, 3, 4, 5]
squared_dict = {x: x**2 for x in nums2}
print(squared_dict)  # Output: {1: 1, 2:        4, 3: 9, 4: 16, 5: 25}
```

```python
nums3 = counter | squared_dict
print(nums3)  
```

```python
murge_dict = {**counter, **squared_dict}
print(murge_dict)
```

```python
from collections import Counter
nums = [2, 2, 2, 3, 3, 4, 4, 5]
lst_counter = Counter(nums)
print(lst_counter)  # Output: Counter({2: 3, 3: 2, 4: 2, 5: 1})
```
