```python
lst = [1, 2, 3, 4, 5]
lst.append("Different data types")
print(lst)
```

```python
# accessing the each element from the list
print(lst[0])  # Accessing the first element
print(lst[1])  # Accessing the second element
print(lst[2])  # Accessing the third element
print(lst[3])  # Accessing the fourth element
print(lst[4])  # Accessing the fifth element
print(lst[5])  # Accessing the sixth element (which is a string)
```

```python
#By writing the for loop we can access the each element from the list
for i in range(len(lst)):
     print(f"index: {i} value: {lst[i]}")
```

```python
#The enumerate() function adds a counter to an iterable and returns it as an enumerate object. This can be used to get both the index and the value of each element in the list.
for i, value in enumerate(lst):
    print(f"index: {i} value: {value}")
```

```python
#I want to print the list range of elements from index 1 to 3
print(lst[0:1])
print(lst[1:2])
print(lst[2:5]) 
print(lst[:-1])#This will print all the elements of the list except the last one
print(lst[1:]) #This will print all the elements of the list starting from index
print(lst[-1: : -1])#This will print the last two elements of the list in reverse order
```

```python
#Modifying the list
lst[0] = 10
print(lst)
```

```python
lst[0:2] = ["New value", "Hari"] #This will replace the last element of the list with "New value"
print(lst)
```

```python
lst[0] = 99
print(lst)
lst[0] = ["New value 1, New value 2"] #This will replace the first element of the list with a new list
print(lst)
```

```python
#List methods
# append() - This method is used to add an element to the end of the list.
lst.append("New element")
#Insert() - This method is used to add an element at a specific index in the list.
lst.insert(1, "Inserted element") #This will insert the element "Inserted element"
#remove() - This method is used to remove the first occurrence of a specified value from the list.
lst.remove("Inserted element") #This will remove the first occurrence of "Inserted element" from the list
print(lst)
#pop() - This method is used to remove and return an element at a given index. If no index is specified, it removes and returns the last item in the list.
lst.pop() #This will remove and return the last item in the list
print(lst)
```

```python
print(lst.count(3)) #This will count the number of occurrences of the value 3 in the list
```

```python
#Reversing the list
lst.reverse() #This will reverse the order of the elements in the list
print(lst)
```

```python
lst.clear() #This will remove all the elements from the list
print(lst)
```

```python
li = [i for i in range(10)]
print(li)
li = [i for i in range(10) if i % 2 == 0] #This will create a list of even numbers from 0 to 9
print(li)
li = [i**2 if i % 2 == 1 else i * 2 if i % 2 == 0 else i for i in range(10)] #This will create a list of squares of numbers from 0 to 9
print(li)
```
