```python

# Tuples are ordered, immutable collections in Python
# They are defined using parentheses () and can store multiple data types

# Creating tuples
empty_tuple = ()
single_item_tuple = (42,)  # Note the comma is required for single item tuples
coordinates = (10, 20)
mixed_tuple = ("Alice", 25, True, 3.14)

# Tuples are immutable - you cannot change their contents after creation
# coordinates[0] = 15  # This would raise a TypeError

# Accessing tuple elements using indexing (0-based)
name = mixed_tuple[0]  # "Alice"
age = mixed_tuple[1]   # 25

# Negative indexing works from the end
last_item = mixed_tuple[-1]  # 3.14

# Tuple slicing
subset = mixed_tuple[1:3]  # (25, True)

# Tuple unpacking - assigning tuple elements to variables
x, y = coordinates
person_name, person_age, is_active, pi_value = mixed_tuple

# Tuples can be used as dictionary keys (unlike lists) because they're immutable
location_data = {(0, 0): "origin", (10, 20): "point A"}

# Common tuple methods
numbers = (1, 2, 3, 2, 4, 2)
count_of_twos = numbers.count(2)  # Returns 3
index_of_three = numbers.index(3)  # Returns 2

# Tuples are useful for returning multiple values from functions
def get_name_age():
    return "Bob", 30

# Tuples support iteration
for item in mixed_tuple:
    print(item)

# Converting between tuples and lists
tuple_from_list = tuple([1, 2, 3, 4])
list_from_tuple = list(coordinates)
```

```python
tpl = ()
#or
tpl = tuple()
#assning the list to a tuple
tpl = tuple([1, 2, 3, 4])
print(tpl)
```

```python
#we can concatenate tuples using the + operator
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
concatenated_tuple = tuple1 + tuple2
print(concatenated_tuple)  # Output: (1, 2, 3, 4, 5, 6)
```

```python
#List elements can be changed but tuple elements cannot be changed
my_list = [1, 2, 3]
my_list[0] = 10  # This is allowed
print(my_list)  # Output: [10, 2, 3]
my_tuple = (1, 2, 3)
#my_tuple[0] = 10  # This will raise a TypeError because tuples are immutable
```

```python
#packing and unpacking tuples
packed_tuple = 1, "hello", 3.14
print(packed_tuple)  # Output: (1, 'hello', 3.14)
```

```python
num, greeting, pi = packed_tuple
print(num)       # Output: 1
print(greeting)  # Output: "hello"
print(pi)        # Output: 3.14
```
