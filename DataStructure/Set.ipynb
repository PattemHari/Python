```python

# Set is a built-in data structure in Python that stores unique, unordered elements
# Sets are mutable (can be modified) and do not allow duplicate values

# Creating sets
empty_set = set()  # Empty set (note: {} creates a dict, not a set)
print(type(empty_set))  # <class 'set'>
numbers = {1, 2, 3, 4, 5}  # Set with initial values
from_list = set([1, 2, 2, 3, 3, 4])  # Creates {1, 2, 3, 4} - duplicates removed

# Adding elements
numbers.add(6)  # Adds single element
numbers.update([7, 8, 9])  # Adds multiple elements

# Removing elements
numbers.remove(1)  # Raises KeyError if element doesn't exist
numbers.discard(10)  # Does not raise error if element doesn't exist
popped = numbers.pop()  # Removes and returns arbitrary element

# Set operations
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

union = set1 | set2  # or set1.union(set2) - all elements from both sets
intersection = set1 & set2  # or set1.intersection(set2) - common elements
difference = set1 - set2  # or set1.difference(set2) - elements in set1 but not set2
symmetric_diff = set1 ^ set2  # or set1.symmetric_difference(set2) - elements in either set but not both

# Set comparisons
is_subset = set1 <= set2  # or set1.issubset(set2)
is_superset = set1 >= set2  # or set1.issuperset(set2)
is_disjoint = set1.isdisjoint(set2)  # True if no common elements

# Common use cases
unique_items = set([1, 1, 2, 2, 3, 3])  # Remove duplicates from list
membership_test = 5 in numbers  # Fast O(1) membership testing
```
