```python

# Lambda functions in Python are anonymous functions that can be defined inline
# They are useful for short, simple functions that don't require a full function definition

# Basic syntax: lambda arguments: expression
# Example of a simple lambda function
square = lambda x: x ** 2

# Lambda functions are commonly used with built-in functions like map(), filter(), and sort()

# Using lambda with map() to apply a function to all items in a list
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))

# Using lambda with filter() to filter items based on a condition
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

# Using lambda with sort() for custom sorting
students = [('Alice', 85), ('Bob', 90), ('Charlie', 78)]
students.sort(key=lambda student: student[1])  # Sort by grade

# Lambda functions can take multiple arguments
add = lambda x, y: x + y
result = add(5, 3)

# Lambda functions can be used in list comprehensions and other contexts
# They are limited to single expressions and cannot contain statements like print or return

# Example of lambda in a conditional expression
max_value = lambda a, b: a if a > b else b

# Lambda functions are particularly useful for short callback functions
# and when you need a simple function for a brief period
```

```python
def add(a, b):
    return a + b
#The above function can be rewritten using a lambda function as follows:
addition = lambda a, b: a + b
#both the add function and the addition lambda function will give the same result when called with the same arguments.
print(addition(5, 3))  # Output: 8
print(add(5, 3))  # Output: 8

# In lambda we should write only one expression and it will return the value of that expression. We cannot use multiple statements in a lambda function.
```
