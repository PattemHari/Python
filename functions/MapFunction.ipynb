```python
#map function is used to apply a specific function to each item of an iterable (like a list) and return a map object (which is an iterator). The syntax for the map function is as follows:
# map(function, iterable)
#Here, function is the function that you want to apply to each item of the iterable, and iterable is the iterable (like a list) that you want to process.
#Example:
def square(x):  
    return x * x
lst = [1, 2, 3, 4, 5]
squared_lst = map(square, lst)
print(list(squared_lst))  # Output: [1, 4, 9, 16, 25]
```

```python
fruits = ['apple', 'banana', 'cherry']
upper_Case_fruits = map(str.upper, fruits)
print(list(upper_Case_fruits))
```
